```python
# 테스트 코드

import torchaudio
import pesto

x, sr = torchaudio.load("vocal.wav")
x = x.mean(dim=0)  # PESTO takes mono audio as input

timesteps, pitch, confidence, activations = pesto.predict(x, sr)
```

```python
# 테스트 코드

import numpy as np
import matplotlib.pyplot as plt

# 오디오 파일의 총 길이 (초)
audio_length = x.shape[-1] / sr

# timesteps를 numpy 배열로 변환하고, 원본 오디오 길이에 맞게 스케일링
scaled_timesteps = timesteps.numpy() * (audio_length / timesteps[-1].item())

# 무음 구간 임계값 설정 (예: confidence 값이 0.9 이하일 때 무음으로 간주)
silence_threshold = 0.95

# 공백 처리된 피치 배열 생성
processed_pitch = np.array([
    pitch[i] if confidence[i] >= silence_threshold else 0  # 무음 구간은 피치를 0으로 설정
    for i in range(len(pitch))
])

# 피치 그래프 그리기 (공백 처리 후)
plt.figure(figsize=(10, 5))
plt.plot(scaled_timesteps, processed_pitch, label="Processed Pitch", color="blue")
plt.xlabel("Time (s)")
plt.ylabel("Pitch (Hz)")
plt.title("Pitch over Time (Processed)")
plt.legend()
plt.grid(True)
plt.show()

import librosa

def freq_to_note(freq):
    if freq == 0:
        return "Silence"
    
    note_number = round(12 * np.log2(freq / 440.0) + 69)  # MIDI 넘버 변환
    note_name = librosa.midi_to_note(note_number)  # 음정 변환
    return note_name

# 유효한 피치 값만 필터링 (0 제외)
valid_pitches = processed_pitch[processed_pitch > 0]

if len(valid_pitches) > 0:
    avg_pitch = np.mean(valid_pitches)  # 평균 피치 계산
    note = freq_to_note(avg_pitch)  # 음정 변환
    print(f"평균 피치: {avg_pitch:.2f} Hz, 음정: {note}")
else:
    print("유효한 피치가 없습니다.")

```

```python
import os
import csv
import torchaudio
import pesto
import numpy as np
import librosa

# 폴더 경로 설정
folder_path = "../vocals/popular_100"  # 분석할 wav 파일들이 있는 폴더 경로
output_csv = "../avg_note/popular_100.csv"

def freq_to_note(freq):
    if freq == 0:
        return "Silence"
    note_number = round(12 * np.log2(freq / 440.0) + 69)
    return librosa.midi_to_note(note_number)

def analyze_pitch(file_path):
    x, sr = torchaudio.load(file_path)
    x = x.mean(dim=0)  # 모노 변환
    
    timesteps, pitch, confidence, _ = pesto.predict(x, sr)
    
    # 오디오 파일의 길이에 맞게 timesteps 스케일링
    audio_length = x.shape[-1] / sr
    scaled_timesteps = timesteps.numpy() * (audio_length / timesteps[-1].item())
    
    # 무음 구간 임계값 설정
    silence_threshold = 0.95
    processed_pitch = np.array([
        pitch[i] if confidence[i] >= silence_threshold else 0
        for i in range(len(pitch))
    ])
    
    # 유효한 피치 값만 필터링
    valid_pitches = processed_pitch[processed_pitch > 0]
    if len(valid_pitches) > 0:
        avg_pitch = np.mean(valid_pitches)
        note = freq_to_note(avg_pitch)
    else:
        avg_pitch, note = 0, "Silence"
    
    return avg_pitch, note

data = []

# 폴더 내 모든 wav 파일 처리
for file in os.listdir(folder_path):
    if file.endswith(".wav"):
        file_path = os.path.join(folder_path, file)
        avg_pitch, note = analyze_pitch(file_path)
        data.append([file, note])
        print(file, note)

# CSV 파일 저장
with open(output_csv, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["title", "note"])
    writer.writerows(data)
```
